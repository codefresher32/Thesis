```python
subset['Start_Time'] = pd.to_datetime(subset['Start_Time'], errors = 'coerce')
subset['Time'] = (subset['Start_Time'].dt.hour % 24) * 60 + subset['Start_Time'].dt.minute % 60
del(subset['Start_Time'])
subset
```
